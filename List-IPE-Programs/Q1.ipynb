```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Convert this file into pandas Data Frame and Display basic statistics like mean, std, quartiles, etc. for this data frame.

df = pd.read_csv('assests/data.csv')

# Display basic statistics
print("Basic Statistics:")
print(df.describe())
```

```python
# 2. Print first and last 5 rows. Also print the shape of the dataframe.

print("Shape of the DataFrame:")
print(df.shape)

print("First 5 rows:")
df.head()
```

```python
# 3. Create a correlation table for the data frame and comment about what kind of correlation is there between Duration and Calories?

correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# 4. Find whether there any null or NA values, drop all such rows if found in the data frame and print the shape of the data frame after dropping.

print("Checking for null or NA values:")
print(df.isnull().sum())

print("Dropping null or NA values:")
df.dropna(inplace=True)

print("Shape of the DataFrame after dropping null or NA values:")
print(df.shape)
```

```python
# 5. Prepare a scatter matrix for the following data frame.

plt.figure(figsize=(10, 8))
pd.plotting.scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal='kde')
plt.show()
```

```python
# 6. Prepare a parallel coordinates for Duration v/s Pulse, Maxpulse and Calories (all 3 other columns).

pd.plotting.parallel_coordinates(df, 'Duration', color=['b', 'g', 'r', 'c', 'm', 'y', 'k'], linewidth=1.5)
plt.title('Parallel Coordinates for Duration vs Pulse, Maxpulse, and Calories')
plt.show()
```

```python
# 7. Prepare a cross-tabulation for Duration v/s Pulse.

cross_tab = pd.crosstab(df['Duration'], df['Pulse'])
print("Cross-tabulation for Duration vs Pulse:")
cross_tab
```

```python
# 8. Do Maxpulse have any outliers? Find using function.

print("Checking for outliers in Maxpulse column:")
def check_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers

outliers_maxpulse = check_outliers(df['Maxpulse'])
print("Outliers in Maxpulse:")
print(outliers_maxpulse)
```
